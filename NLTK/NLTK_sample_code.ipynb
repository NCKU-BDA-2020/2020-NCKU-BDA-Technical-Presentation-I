```python
from nltk.corpus import stopwords
print(stopwords.words('english'))
```

```python
def content_fraction(text):
    stopwords=nltk.corpus.stopwords.words('english')
    content=[w for w in text if w not in stopwords]
    return len(content)/len(text)
content_fraction(brown.words(fileids='cc04'))
```

```python
from nltk.corpus import wordnet as wn
wn.synsets('motorcar')
wn.synsets('trunk')
```

```python
wn.synset('car.n.01').lemma_names()
for synset in wn.synsets('trunk'):
    print(synset.lemma_names())
```

```python
wn.synset('car.n.01').definition()
wn.synset('trunk.n.01').definition()
```

```python
#上位詞
motorcar = wn.synset('car.n.01')
types_of_motorcar = motorcar.hypernyms()
#下位詞
otorcar = wn.synset('car.n.01')
types_of_motorcar = motorcar.hyponyms()
sorted(lemma.name() for synset in types_of_motorcar for lemma in synset.lemmas())
```

```python
motorcar = wn.synset('car.n.01')
motorcar.hypernym_paths()
motorcar.root_hypernyms()
```

```python
# 露脊鯨 vs 虎鯨
right = wn.synset("right_whale.n.01") 
orca = wn.synset("orca.n.01")
right.lowest_common_hypernyms(orca)
# 露脊鯨 vs 陸龜
right = wn.synset("right_whale.n.01") 
tortoise = wn.synset("tortoise.n.01")
right.lowest_common_hypernyms(tortoise)
# 露脊鯨 vs 小說
right = wn.synset("right_whale.n.01") 
novel = wn.synset("novel.n.01")
right.lowest_common_hypernyms(novel)
```

```python
right.path_similarity(orca)      #露脊鯨和虎鯨
right.path_similarity(tortoise)  #露脊鯨和陸龜
```
